Zunächst definieren wir eine Funktion, die zu gegebenen Stützstellen das $j$-te Lagrange-Polynom für eine Liste an Punkten berechent.

```python
def lagrange_polynomial(j, nodes, points):
    '''
        Auswertung des j-ten Lagrange-Polynoms zu den Stützstellen nodes
        Parameter:
            j:             Grad des Polynoms
            nodes:         Stützstellen der Interpolation als np.array
            points:        Auswertungspunkte als np.array
        Rückgabewert: 
            values:        Funktionswerte des Polynoms an den Auswertungspunkten als np.array
    '''
    n = nodes.shape[0]

    values = np.ones_like(points)

    for i in range(n):
        if i != j:
            values *= (points - nodes[i]) / (nodes[j] - nodes[i])

    return values
```

Wir können diese Funktion nun dazu nutzen, die Lagrange-Polynome für gegebene Stützstellen zu plotten. Führen Sie dazu die folgene Zelle aus, und beschreiben Sie kurz, was die Funktionswerte der Polynome in den Stützstellen erfüllen.

```python
nodes = np.array([-1, 0, 1, 2, 3.5, 5])
points = np.linspace(-1.1, 5.1, 200)

plt.figure()
for i in range(nodes.shape[0]):
    values = lagrange_polynomial(i, nodes, points)
    plt.plot(points, values)
plt.plot(nodes, np.zeros_like(nodes), 'ro')
plt.plot(nodes, np.ones_like(nodes),  'bo')
plt.show()
```

Verwenden Sie nun Ihre Implementierung der Lagrange Polynome, um zu gegebenen Stützstellen und Funktionswerten das Interpolationspolynom an vorgegebenen Werten zu berechnen.

```python
def interpolation(function_values, nodes, points):
    '''
        Auswertung des Interpolationspolynoms
        Parameter:
            fuction_values: Funktionswerte an den Stützstellen als np.array
            nodes:          Stützstellen der Interpolation als np.array
            points:         Auswertungspunkte als np.array
        Rückgabewert: 
            values:         Funktionswerte des Polynoms an den Auswertungspunkten als np.array
    '''
    n = nodes.shape[0]

    values = np.zeros_like(points)

    for i in range(n):
        values += function_values[i] * lagrange_polynomial(i, nodes, points)

    return values
```

Wenn Sie mit verschiedenen Stützstellen experimentieren, werden Sie feststellen, dass die Interpolation, abhängig von der Position der Stützstellen, oft relativ große Fehler macht. Wir wollen das nun etwas genauer untersuchen.

Dafür interpolieren wir die **Runge-Funktion** $f(x) = \frac{1}{1+x^2}$ sowohl an *äquidistanten* Stützstellen, als auch an den Nullstellen der Tschebyscheff-Polynome. Führen Sie dazu die nächste Zelle aus und beschreiben sie darunter kurz, wie sich die beiden Interpolationspolynome verhalten. 

```python
n = 11 # Anzahl Stüzstellen
f = lambda x: 1/(1+x**2)

points = np.linspace(-5, 5, 500)

equidistant_nodes = np.linspace(-5, 5, n)
tschebyscheff_nodes = 5 * np.cos( (2 * np.arange(n) + 1) / (2 * n) * np.pi )

plt.figure()
plt.plot(points, f(points), linewidth=3, color='black', label="Runge-Funktion")
plt.plot(equidistant_nodes, f(equidistant_nodes), 'ro')
plt.plot(points, interpolation(f(equidistant_nodes), equidistant_nodes, points), 'r--', label='Äquidistante Interpolation')
plt.plot(tschebyscheff_nodes, f(tschebyscheff_nodes), 'bo')
plt.plot(points, interpolation(f(tschebyscheff_nodes), tschebyscheff_nodes, points), 'b--', label='Tschebyscheff Interpolation')
plt.legend(loc='upper right')
```
